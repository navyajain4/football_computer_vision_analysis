## Importing Necessary Libraries

Here we use YOLOv8 to conduct analysis. For that, we import YOLO from `ultralytics` and run the model to see what all objects it is detecting.

```python
from ultralytics import YOLO
```

```python
model = YOLO('yolov8l')
```

```python
result = model.predict('/content/08fd33_4.mp4', save = True)
```

```python
print(result[0])
```

```python
print('====================')
for box in result[0].boxes:
  print(box)
```
